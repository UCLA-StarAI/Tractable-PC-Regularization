```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
pc_names = []
overfit_degree = []
pseudocount_res3 = []
with open("pseudocount_logfile.txt", "r") as f:
    for line in f.readlines():
        pc_names.append(line.split(" ")[1])
        pseudocount_res3.append(float(line.split(" ")[2]))
        overfit_degree.append(float(line.split(" ")[3]))
        
overfit_degree = np.array(overfit_degree)
pseudocount_res3 = np.array(pseudocount_res3)
```

```python
soft_reg_res = np.zeros([pseudocount_res.size])
with open("soft_reg_logfile.txt", "r") as f:
    for line in f.readlines():
        pc_name = line.split(" ")[1]
        idx = pc_names.index(pc_name)
        soft_reg_res[idx] = float(line.split(" ")[2])
```

```python
ent_reg_res3 = np.zeros([pseudocount_res.size])
with open("entropy_reg_logfile-new3.txt", "r") as f:
    for line in f.readlines():
        pc_name = line.split(" ")[1]
        idx = pc_names.index(pc_name)
        ent_reg_res3[idx] = float(line.split(" ")[2])
        
ent_reg_res3 = np.array(ent_reg_res3)
```

```python
ent_soft_reg_res3 = np.zeros([pseudocount_res.size])
with open("ent_soft_reg_logfile-new2.txt", "r") as f:
    for line in f.readlines():
        pc_name = line.split(" ")[1]
        try:
            idx = pc_names.index(pc_name)
            ent_soft_reg_res3[idx] = float(line.split(" ")[4])
        except ValueError:
            pass
        
ent_soft_reg_res3 = np.array(ent_soft_reg_res3)
```

```python
import scipy.io as sio
sio.savemat("export/results.mat",
            {
                "overfit_degree": overfit_degree,
                "pseudocount_res": pseudocount_res3,
                "soft_reg_res": soft_reg_res,
                "ent_reg_res": ent_reg_res3,
                "ent_soft_reg_res": ent_soft_reg_res3
            })
```

```python
for name, pseudo, soft, ent, se in zip(pc_names, pseudocount_res3, soft_reg_res, ent_reg_res3, ent_soft_reg_res3):
    if "" in name:
        print(name, pseudo, soft, ent, se)
        print()
```
